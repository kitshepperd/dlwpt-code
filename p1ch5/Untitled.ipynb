```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3,1000)
```

```python
A = np.tanh(-2*x - 1.25)
B = np.tanh(x + 0.75)
C = np.tanh(4*x + 1)
```

```python
fig, ax = plt.subplots(1)
ax.plot(x, A, 'r')
ax.plot(x, B, 'b')
ax.plot(x, C, 'g')
ax.set_ylim(-3,3)
plt.show()
```

```python
C_A_B = np.tanh(4*(A+B) + 1)
```

```python
fig, ax = plt.subplots(1)
ax.plot(x, C_A_B, 'r')
ax.set_ylim(-3,3)
plt.show()
```
